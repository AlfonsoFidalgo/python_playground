```python
df = pd.read_excel('dsa_es_cat.xlsx', header=1) #header indicate the row where they are
```

# Code to analyse categories

```python
def parseCat (str):
    if type(str) != float:
        arr = str.split('\n')
        return arr[0]
    else:
        return "No Cat"
```

```python
df['Cat1'] = df['Categories'].apply(parseCat)
```

```python
categories = df[['Cat1', 'Impressions', 'Clicks', 'Cost', 'Conversions']]
```

```python
summary = categories.groupby('Cat1').sum()
```

```python
summary['CPA'] = summary['Cost']/summary['Conversions']
```

```python
summary[summary['Conversions'] > 0].sort_values(by='Conversions', ascending=False)
```

```python
summary[summary['Conversions'] == 0].sort_values(by='Cost', ascending=False)
```
