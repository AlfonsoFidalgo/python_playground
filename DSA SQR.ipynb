```python
def parseUrl (url):
    arr = url.split('/')
    if len(arr[len(arr)-1]) == 1:
        return arr[len(arr)-2]
    else:
        return arr[len(arr)-1]
```

```python
def categorise (url):
    if '-ttd' in url:
        return 'TTD'
    elif '-t' in url:
        return 'Recommendation'
    elif '-g' in url:
        return 'Group/Category'
    elif '-a' in url:
        return 'Attraction'
    elif '-' not in url:
        return 'TTD'
    else:
        return 'Product'
```

```python
df['URLtype'] = df['URL'].apply(parseUrl)
```

```python
df['URL Cat'] = df['URLtype'].apply(categorise)
```

```python
sqr = df[['Search term','Impressions', 'Clicks' ,'Conversions', 'Cost', 'URL', 'URL Cat']]
```

```python
sqr.head()
```

```python
summary = sqr.groupby('URL Cat').sum()
```

```python
summary['CPA'] = summary['Cost']/summary['Conversions']
```

```python
summary['CVR'] = (summary['Conversions']/summary['Clicks'])*100
```
