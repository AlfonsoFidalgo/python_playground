```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('mkl.csv')
```

```python
df['urltype'].unique()
```

```python
def brandGen(type):
    if 'Brand' in type:
        return 'Brand'
    else:
        return 'Non Brand'
```

```python
df[df['urltype'].apply(lambda x: 'Brand' not in x)]
```

```python
df['Brand/Non Brand'] = df['urltype'].apply(brandGen)
```
