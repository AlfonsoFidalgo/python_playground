```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel('dsa.xlsx')
```

```python
df['Account'].unique()
```

```python
df_new = df[['DayCol', 'Account', 'Conversions', 'Revenue', 'Cost']].groupby(by=['Account', 'DayCol']).sum()
```

```python
df_new.head(15)
```

```python
df_new.reset_index(inplace=True)
```

```python
df_new.head()
```

```python
df_new['ROAS'] = df_new['Revenue'] / df_new['Cost']
df_new['CPA'] = df_new['Cost'] / df_new['Conversions']
```

# SPAIN

```python
df_es = df_new[(df_new['Account'] == 'Viator - Spanish')]
```

```python
sns.set_style("whitegrid")
lm = sns.lmplot('ROAS', 'CPA', df_es, sharex=False, sharey=False, size=8, aspect=1.5, fit_reg=True, ci=95)
lm.ax.set_ylim(15,90)
lm.ax.set_xlim(0.4,None)
#axes = lm.axes
#axes[0,0].set_xlim(0,6)
#axes[0,0].set_ylim(10,60)
#axes[0,1].set_ylim(10,)
```

```python
#pd.rolling_sum(df_es['Cost'], 7)
cost7 = df_es['Cost'].rolling(window=7, center=False).sum()
rev7 = df_es['Revenue'].rolling(window=7, center=False).sum()
conv7 = df_es['Conversions'].rolling(window=7, center=False).sum()
df_es['avg ROAS'] = rev7 / cost7
df_es['avg CPA'] = cost7 / conv7
```

```python
plt.figure(figsize=(12,8))

sns.tsplot(df_es['avg ROAS'].dropna())
#sns.tsplot(df_es['ROAS'])

#sns.tsplot(df_es['avg CPA'])
```

```python
plt.figure(figsize=(12,8))
sns.tsplot(df_es['avg CPA'].dropna())
#sns.tsplot(df_es['CPA'])
```

# Spain by device

```python
df_es_device = df[(df['Account'] == 'Viator - Spanish') & (df['Cost'] > 0)][['DayCol', 'Revenue', 'Cost', 'Conversions', 'Device', 'Clicks']].groupby(by=['DayCol', 'Device']).sum()
```

```python
df_es_device.reset_index(inplace=True)
```

```python
df_es_device['CPA'] = df_es_device['Cost'] / df_es_device['Conversions']
df_es_device['ROAS'] = df_es_device['Revenue'] / df_es_device['Cost']
df_es_device['CVR'] = df_es_device['Conversions'] / df_es_device['Clicks']
```

```python
df_es_device.head(10)
```

```python
lm = sns.lmplot(x='ROAS', y='Revenue', data=df_es_device, hue='Device', size=8, aspect=1.5)
lm.ax.set_ylim(0,2000)
lm.ax.set_xlim(0,None)
```

```python
sum_device_es = df_es_device[['Device', 'Revenue', 'Cost', 'Conversions', 'Clicks']].groupby(by=['Device']).sum()
```

```python
sum_device_es['Rev per Conv'] = sum_device_es['Revenue'] / sum_device_es['Conversions']
sum_device_es['Rev per Click'] = sum_device_es['Revenue'] / sum_device_es['Clicks']
sum_device_es['ROAS'] = sum_device_es['Revenue'] / sum_device_es['Cost']
```
