```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('daily_performance.csv', names=['Day', 'AdGroup', 'Campaign', 'Device', 'Revenue', 'Conversions', 'Cost', 'Impressions', 'Clicks'])
```

```python
df.head()
```

```python
df[df['Campaign'].apply(lambda x: 'Brand' in x)].head()
```

```python
def catCamp(camp):
    if 'Brand' in camp:
        return 'Brand'
    elif 'DSA' in camp:
        return 'DSA'
    else: return 'Generic'
```

```python
def catAg(ag):
    if 'viator' in ag:
        return 'Brand'
```

```python
df['Type'] = df['AdGroup'].apply(catAg)
```

```python
df['Type'] = df['Campaign'].apply(catCamp)
```

```python
df[(df['AdGroup'].apply(lambda x: 'viator' in x)) & (df['Type'] != 'Brand')].head(25)
```

```python
df[(df['AdGroup'].apply(lambda x: 'viator' in x)) & (df['Type'] != 'Brand')]
```

```python
df.loc[(df['AdGroup'].apply(lambda x: 'viator' in x)) & (df['Type'] != 'Brand'), 'Type'] = 'Brand'
```

```python
df.head(5)
```

```python
df2 = df[['Day', 'Revenue', 'Conversions', 'Cost', 'Impressions', 'Clicks', 'Type']].groupby(by=['Type', 'Day']).sum()
```
