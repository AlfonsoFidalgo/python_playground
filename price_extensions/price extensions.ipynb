```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('Final URL report.xlsx', header=1)
```

```python
df.dropna(inplace=True)
```

```python
df['URL Clean'] = df['Final URL'].apply(lambda x: x.split('?')[0])
```

# Helper functions

```python
def parseUrl (url):
    arr = url.split('/')
    if len(arr[len(arr)-1]) == 1:
        return arr[len(arr)-2]
    else:
        return arr[len(arr)-1]
```

```python
def categorise (url):
    if '-ttd' in url:
        return 'TTD'
    elif '-t' in url:
        return 'Recommendation'
    elif '-g' in url:
        return 'Group/Category'
    elif '-a' in url:
        return 'Attraction'
    elif '-' not in url:
        return 'TTD'
    else:
        return 'Product'
```

```python
def parseCat (str):
    if type(str) != float:
        arr = str.split('\n')
        return arr[0]
    else:
        return "No Cat"
```

# End of functions

```python
df['URL Cat'] = df['URL Clean'].apply(parseUrl)
```

```python
df['Category'] = df['URL Cat'].apply(categorise)
```

```python
df['Destination'] = df['URL Cat'].apply(lambda x: x.split('-')[0])
```

```python
df['Attraction'] = df[df['Category'] == 'Attraction']['URL Cat'].apply(lambda x: x.split('-')[1])
```

```python
df['Conversions'] = df['Conversions'].apply(lambda x: int(x.split('.')[0].replace(',','')))
```
