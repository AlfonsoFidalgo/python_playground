```python
import numpy as np
import pandas as pd
```

```python
from lxml import html
import requests
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sl = pd.read_excel('Book1.xlsx')
```

```python
sl.head()
```

```python
sl['Final URL EN'] = sl['Final URL'].apply(lambda x: x.replace('es.viator.com/es/7380', 'www.viator.com'))
```

```python
def getPrice(x):
    page = requests.get(x)
    tree = html.fromstring(page.content)
    price = tree.xpath('//span[@class="price-amount"]/text()')
    price_float = []
    for i in price:
        try:
            price_float.append(float(i))
        except:
            print('something wrong here')
    try:
        return min(price_float)
    except:
        return 'Couldnt get price'
```

```python
sl['Final URL EN'].nunique()
```

```python
prices = []
```

```python
for i in sl['url']:
    min_price = getPrice(i)
    prices.append([i, min_price])
```

```python
prices
```

```python
prices_df = pd.DataFrame(prices, columns=['URL', 'Price from USD'])
```
