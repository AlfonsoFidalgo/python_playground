```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('ES_price_exten_dario.xlsx')
```

```python
campaigns = df['Campaign'].unique()
```

output = []
for i in campaigns:
    camp = df[df['Campaign'] == i].drop('Campaign', axis=1)
    camp.reset_index()
    #camp.drop('index', axis=1, inplace=True)
    row = [i]
    for j in range(0, len(camp)):
        row.append(camp.loc[j].values)
    output.append(row)

```python
count = 0
current = df.loc[0]['Campaign']
result = []
for i in range(0, len(df['Campaign'])):
    if df.loc[i]['Campaign'] == current:
        result.append(count)
        count = count + 1
    else:
        count = 0
        current = df.loc[i]['Campaign']
        result.append(count)
        count = count + 1
```

```python
df['PE Num'] = result
```
